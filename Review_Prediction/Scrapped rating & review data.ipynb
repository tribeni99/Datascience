```python
import pandas as pd
import selenium
import time
from selenium import webdriver

import warnings
warnings.filterwarnings("ignore")

driver=webdriver.Chrome(r"C:\\Users\\p.c\\Downloads\\chromedriver_win32\\chromedriver.exe")
```

```python
items=['Mobiles','headphones','smartwatches','laptops','home theatre','printers','Professional cameras']
ratings_=[]
reviews_=[]
for item in items:
    driver.get("https://www.amazon.in/")
    driver.find_element_by_id("twotabsearchtextbox").send_keys(item)
    driver.find_element_by_id("nav-search-submit-button").click()
    links=[]
    links.extend([w.get_attribute('href') for w in driver.find_elements_by_xpath("//a[@class='a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal']")])
    

    for i in links[1:16]:
        driver.get(i)
        try:
            driver.find_element_by_xpath("//a[@class='a-link-emphasis a-text-bold']").click()
        except:
            continue
        counter="y"
        while counter=='y':
            try:
                reviews=[]
                ratings=[]
                ratings.extend([r.get_attribute("title") for r in driver.find_elements_by_xpath("//a[@class='a-link-normal']")][13:33:2])
                reviews.extend([r.text for r in driver.find_elements_by_xpath("//a[@class='a-size-base a-link-normal review-title a-color-base review-title-content a-text-bold']/span")])
                try:
                    driver.find_element_by_xpath("//li[@class='a-last']").click()
                except:
                    counter='n'
                if len(ratings)==len(reviews):
                    ratings_.extend(ratings)
                    reviews_.extend(reviews)
        
            except:
                continue
```
