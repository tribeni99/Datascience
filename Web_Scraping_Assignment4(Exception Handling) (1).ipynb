1. Scrape the details of most viewed videos on YouTube from Wikipedia.
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
# Activating the chrome browser
driver=webdriver.Chrome(r"C:/Users/tribe/Downloads/chromedriver_win32/chromedriver.exe")
#Loading url to the chrome driver
url = 'https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
driver.get(url)
```

```python
driver.maximize_window()
```

```python
#creating list for rank, uploader)name, video name, no of views, date of upload
rank = []
video_names = []
artist = []
views = []
upload_dates = []
```

```python
#getting ranks of the videos
try:
    rank_obj = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[1]')[:30]
    rank_obj
    for i in rank_obj:
        rank.append(i.text.replace('.',''))
except NoSuchElementException:
    rank.append('-')
print(rank)
print("++++++++++++++++++++++++++++++++++")
#getting names of the videos
try:
    videoname_obj = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[2]')[:30]
    videoname_obj
    for i in videoname_obj:
        video_names.append(i.text)
except NoSuchElementException:
    video_names.append('-')
print(video_names)
print("++++++++++++++++++++++++++++++++++")
#getting names of artist
try:
    uploader_obj = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[3]')[:30]
    for i in uploader_obj:
        artist.append(i.text)
except NoSuchElementException:
    artist.append('-')
print(artist)
print("++++++++++++++++++++++++++++++++++++++++")
#getting numbers of views of the videos
try:
    view_obj = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[4]')[:30]
    view_obj
    for i in view_obj:
        views.append(i.text)
except NoSuchElementException:
    views.append('-')
print(views)
print("++++++++++++++++++++++++++++++++++++++")
#getting upload date
try:
    date_obj = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[5]')[:30]
    date_obj
    for i in date_obj:
        upload_dates.append(i.text)
except NoSuchElementException:
    upload_dates.append('-')
print(upload_dates)
print("+++++++++++++++++++++++++++++++++++++++++")
    
        
```

```python
print(len(views),len(rank),len(video_names),len(artist),len(upload_dates))
```

```python
#Since the length are same,we can creat dataframe
WIKI_VIEWS=pd.DataFrame({})
WIKI_VIEWS['RANK']=rank
WIKI_VIEWS['VIDEOS_NAME']=video_names
WIKI_VIEWS['ARTIST']=artist
WIKI_VIEWS['VIEWS']=views
WIKI_VIEWS['UPLOAD_DATE']=upload_dates
WIKI_VIEWS.reset_index(drop=True,inplace=True)
WIKI_VIEWS.index+=1
WIKI_VIEWS
```
